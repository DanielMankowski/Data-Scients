```python
%matplotlib inline

import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default') # haciendo los graficos un poco mas bonitos xD
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
# Levantamos los datos obtenidos de Properati
propiedades = pd.read_csv('./datasets/properati-AR-2017-08-01-properties-sell.csv',sep=',',error_bad_lines=False,low_memory=False)
propiedades = propiedades[propiedades['state_name'].str.contains('Capital Federal') | \
                            propiedades['state_name'].str.contains('G.B.A.')]

propiedades = propiedades.loc[:,['id' ,  
                                 
                              'lat-lon' , 'lat' , 'lon' , 'price' , 'surface_total_in_m2' , 'surface_covered_in_m2' ,  
                               'price_usd_per_m2' ]]
propiedades.dropna(inplace=True)
# propiedades.dropna(inplace=True)
propiedades = propiedades[(propiedades['price_usd_per_m2'] > 300) & ( propiedades['price_usd_per_m2'] < 10000 )]
propiedades = propiedades[(propiedades['price'] > 45000) & ( propiedades['price'] < 300000 )]
propiedades.info()
propiedades.describe()
```

```python
# Levantamos los datos a predecir de Properati
prop_test = pd.read_csv('./datasets/properati_dataset_testing_noprice.csv',sep=',',error_bad_lines=False,low_memory=False)
prop_test.info()
```

```python
prop_test.loc[:, 'state_name'].value_counts()
```

```python
# test youtube
import scipy
import urllib
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn import neighbors
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split

from sklearn import metrics
```

```python
x_prime = propiedades.ix[:,{2,3,4,5,6,7}]

y = propiedades.ix[:,4]
# neigh = KNeighborsClassifier(n_neighbors=3)
```

```python
x = preprocessing.scale(x_prime)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size = .10, random_state = 17 ) 
```

```python
clf = neighbors.KNeighborsClassifier()
clf.fit(x_train, y_train)
```

```python
y_expect = y_test
y_pred = clf.predict(x_test)
print(metrics.classification_report(y_expect, y_pred))
```
